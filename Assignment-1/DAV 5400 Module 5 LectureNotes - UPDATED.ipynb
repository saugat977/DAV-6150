## Exploratory Data Analysis (EDA)

- EDA = Analyzing a data set for purposes of summarizing its characteristics, identifying relationships between its attributes, and discovering patterns, trends, outliers, missing values and invalid values within the data.


- Thorough EDA work is crucial for purposes of making sense of a data set and helping you to determine what questions you can realistically expect to ask and answer using the contents of a data set.


- __A concise explanatory narrative of your EDA findings is *ALWAYS* necessary: you need to be able to clearly communicate your EDA work to others__. If you can't clearly articulate your EDA findings, how can anyone else validate the work you've done?


## Univariate Analysis

The first step in any EDA should be the analysis of the distribution of each individual attribute contained within the data set(s) you intend to work with. (__NOTE__: Attributes that you have no intention of utilizing for any purpose can be ignored during this step, though you should provide a reader of your work with a clear explanation as to why any such attributes are not planned to be used).

### Missing / Invalid Data Values

- For each attribute: take a count of the number of values that either don't conform to the attribute's known valid value range and/or that are missing in their entirety (e.g., how many 'NaN' values are found for a given attribute? What about other 'NaN' proxies, e.g., '?', 'unk', etc.)

- What percentage of an attribute's values are missing/unknown?


### Use of Summary Statistics

For each attribute, we must provide relevant summary statistics.

- __For each numeric attribute__: max, min, mean, median, interquartile range bounds, numeric value range bounds, etc.


- __For each categorical attribute__: How many valid categorizations are there for a given attribute within the data set? (NOTE: It might prove to be either more than or less than the number of categories specified within any data dictionary you have been provided); How are those categorizations identified within the data? What is the frequency/count of each possible categorization within the data set? Identify the most frequently occurring data value within the category (often referred to as the __mode__ value); 


### Use of Appropriate Graphics

For each attribute, we must provide appropriate graphical representations of the distribution of the attribute.

- __For numeric data__: provide __both a histogram AND a box plot__; For numeric data that has a chronology (e.g., time series data), we should also include a line plot that shows how the value of the attribute changes over time.


- __For categorical data__: use of __bar plots__. Note that pie charts should __NEVER__ be used within the context of professional-caliber analytics work due to the ease with which they can be misinterpreted.


### Discovering Patterns & Trends

- For data that has a chronology, what (if any) repeating patterns can be identified? 


- For data that has a chronology, is there a directional tendency evident?


### Outliers

- For each attribute (both numeric and categorical): are there data values that fall far outside the norm for the data? 


- For numerics: use scatter plots + box plots to find outliers


- For categoricals: check for items that either don't appear to conform to the known valid values for the attribute OR occur very infrequently within the attribute.

## Bivariate / Multi-variate Analysis

After analyzing the distribution of each individual attribute, we then explore potential inferential/predictive relationships __between two or more variables__.

### Identifying Relationships Between Pairs of Attributes

- For __pairs of numeric attributes__: Calculate correlation metrics (use of a correlation matrix might be appropriate), scatter plots; 


- For __pairs of categorical variables__: Use a __grouped bar chart__", wherein the __frequency counts__ of categorical values from one variable are plotted against the categories present within another categorical variable.  For examples, see: https://thinkingneuron.com/how-to-visualize-the-relationship-between-two-categorical-variables-in-python/


- For a __categorical/numerical pair of variables__: Use of "__pair plots__", e.g., strip plots, swarm plots, category-specific box plots or barplots, i.e., the numeric values that align with each categorical value are captured within a separate element of a single graphic. For examples, see: http://seaborn.pydata.org/tutorial/categorical.html 


### Creating a Correlation Matrix

- In Pandas, __after__ isolating the numeric attributes within a dataframe, use df.corr(): https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.corr.html


- __VERY IMPORTANT__: Make sure that any categorical content is __excluded__ from the input to your correlation matrix !!


- A correlation matrix can help you understand the relationships between __numeric__ variables within a data set.


- Positive numbers indicate a positive relationship between the attributes (i.e., as the values of one attribute increase, the values of the second attribute are likely to increase).


- Negative numbers indicate a negative relationship between the attributes (e.g., as the values of one attribute increase the values of the second attribute are likely to decrease).


- The larger the correlation value, the greater the degree of correlation (either positive or negative) between the variables.


- Correlation metrics are not proof of a __meaningful__ correlation between two attributes: __correlation does not imply "causation"__!!! It is possible to find meaningless correlations between all sorts of data values.


- Correlation metrics can provide clues as to which variable pairs might be worthy of further examination to determine whether or not they have a __meaningful__ correlation.


- Even relatively low correlation metrics can be indicative of a meaningful directional relationship between two attributes/variables.


#### Why can't we use correlation metrics to assess relationships between / with categorical variables?

- It is __NEVER__ appropriate to use a correlation calculation to either assess or determine "correlations" between numeric and categorical variables. Correlation equations provide an indication of the strength of a potential linear relationship between a pair of variables. __The "spacing" between categorical classifications has no interval OR ratio meaning__ since __categorical data is QUALITATIVE__ and __NOT QUANTITATIVE__. (For an explanation of nominal vs. ordinal vs. interval vs. ratio, see this link: https://www.graphpad.com/support/faq/what-is-the-difference-between-ordinal-interval-and-ratio-variables-why-should-i-care/#:~:text=right%20statistical%20analysis.-,Nominal,standard%20deviation%2C%20would%20be%20meaningless.). 


- __No meaningful mathematical result can be derived from the addition, subtraction, multiplication, or division of a categorical variable's possible data values__ since the nominally numeric values __are not QUANTITATIVE__.

#### But what if we numerically encode the classifications?

- Numeric encoding of classifications __merely replaces a text-based symbolic category label with digit-based symbolic category label__. Changing the form of the labels __in no way changes the fact that the "spacing" between the classification labels has no interval OR ratio meaning__. Therefore, the results of any "math" we apply to the digit-based labels __are STILL devoid of any interval OR ratio meaning__ since the digits are merely "labels" for __qualitative__ values.


- So while you can obtain an output from a correlation calculation based on numeric encodings of categorical inputs, the output will not have any relevant "meaning" (i.e., garbage in ... garbage out.)

### Length (example of how to analyze a numeric attribute)

```python
pb_df['Length'].describe()
```

__NOTE__: The content of the __Length__ attribute should be numeric. However, the __describe()__ function is treating it as non-numeric. Why is that? 

It appears that there are missing data values within the __Length__ attribute that are indicated by a question mark character ('__?__'). As such, if we wish to make use of the content of the attribute as numeric data, we must convert the entire content of the attribute to numeric format, starting with replacing the __?__ characters with __NaN__:

```python
# check the data type of the Length variable: it is not numeric so we will need to convert it
pb_df["Length"].dtype
```

```python
import numpy as np

# For the Length variable we need to replace the '?' with NaN before doing any computations:
pb_df["Length"].replace({'?':np.nan}, inplace = True)
```

```python
# convert the data type of the Length column to numeric
pb_df["Length"] = pd.to_numeric(pb_df["Length"])
```

Now that the content of the attribute has been converted to numeric format, we can continue with our univariate analysis.

```python
# get summary statistics
pb_df['Length'].describe()
```

%matplotlib notebook

```python
import matplotlib.pyplot as plt
import seaborn as sns

# create a histogram + boxplot for the distribution of 'Length'

# side-by-side boxplot and histogram of one set of data

# define a new figure for plotting
plt.figure(1)

# define the plot that will appear on the lefthand side of the subplot:
# the '121' value passed to the function says: "I am defining a 1x2 arrangement
# of subplots (as indicated by the first 2 digits '12'), and this is the 
# specification for the plot that will appear on the lefthand side (as indicated
# by the last digit '1')"
plt.subplot(121)

# define the type of plot you want to appear
sns.boxplot(pb_df['Length'], orient = 'v');

# give the plot a title
plt.title('Boxplot of Length')

# Here we'll blank out the x-axis label since it isn't really meaningful
# for a box plot
plt.xlabel(' ')

# --------------------------------------------------------------------------
# now define the subplot that will appear on the righthand side of our 1x2
# subplot ('122' = 1x2, position 2)
plt.subplot(122)

# define the type of plot you want to appear: set the number of bins
# for the x axis to 10
sns.distplot(pb_df['Length'], bins = 10)

# give the plot a title
plt.title('Histogram of Length')

plt.show();
```

The summary statistics, histogram, and boxplot show that the distribution of the __Length__ attribute is significantly right-skewed, with a mean value (1567.47) that exceeds the median (1300). The summary statistics also show that a nearly 25% of the observations (i.e., 27/108) within the dataset lack a valid __Length__ value. Both the histogram and the boxplot show evidence of possible outliers, i.e., __Length__ values that fall far beyond the upper whisker of the distribution as shown in the boxplot.

## Bivariate Analysis

Here are some examples of the types of analysis you would want to be sure to include as part of a bivariate analysis for the Pittsburgh Bridges data set:

### Length (numeric) vs.Year Built (numeric)

```python
# define a new figure for plotting
plt.figure()

plt.scatter(x = 'Erected', y = 'Length', data = pb_df)
plt.title("Year Built vs. Length")
plt.show()
```

Plotting __Length__ vs. year built (i.e., __Erected__) shows that up until approximately 1890, nearly all of the bridges represented within the dataset were of lengths that were significantly less than 1500 feet. Starting in approximately 1890, the lengths of bridges being built in the Pittsburgh area increased dramatically, with lengths of more than 2000 feet becoming quite common.

### Length (numeric) vs. Material (categorical)

```python
# define a new figure for plotting
sns.catplot(data=pb_df, x="Material", y="Length")
```

The lengths of all wooden bridges and most iron bridges appear to be 1250 feet or less, while one iron bridge is slightly more than 1500 feet in length. By contrast, steel bridges (which are quite more numerous within the data set) have been used for a wide variety of bridge lengths, ranging from less than 500 feet to more than 4500 feet.

### River (categorical) vs. Purpose (categorial)

```python
CrosstabResult=pd.crosstab(index=pb_df['Purpose'],columns=pb_df['River'])

CrosstabResult.plot.bar(figsize=(7,4), rot=0)
```

```python
CrosstabResult
```

The majority of highway bridges in the Pittsburgh area have been built across the Allegheny river, while the majority of railroad bridges in the area have been built across the Monongahela river. One pedestrian bridge has been built across the Allegheny river, as have four aqueducts.

## Multivariate analysis

### Material vs. Length vs. River

```python
sns.catplot(x='Material', y='Length', hue='River',
               kind='bar', data=pb_df);
```

In the above, we see that, on average, the longest steel bridges in the Pittsburgh area have been constructed across the Ohio river, while, on average, the longest wood and iron bridges have been constructed across the Monongahela river.

### Material vs. Length vs. Purpose

```python
sns.catplot(x='Material', y='Length', hue='Purpose',
               kind='bar', data=pb_df);
```

In the above, we see that, on average, the longest bridges of any type of material have been constructed using steel. Furthermore, we see that highway bridges have been constructed of either wood, iron or steel, while railroad bridges have been constructed using iron or steel.
